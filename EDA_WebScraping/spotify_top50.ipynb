```python
# We are going to grab the songs from the top 50 playlist from Spotify that are on the link below (playlist_link)
playlist_link = "https://open.spotify.com/playlist/37i9dQZEVXbNG2KDcFcKOF?si=1333723a6eff4b7f"
playlist_URI = playlist_link.split("/")[-1].split("?")[0]
print(playlist_URI)
track_uris = [x["track"]["uri"] for x in sp.playlist_tracks(playlist_URI)["items"]]
print(len(track_uris))
print(track_uris)
```

```python
# here we colect the tracks URI's (from the top 50 playlist) and save them on a list (track_uri)
track_uri =[]

for track in sp.playlist_tracks(playlist_URI)["items"]:
    #URI
    track_uri.append(track["track"]["uri"])
    
    #Track name
    track_name = track["track"]["name"]
    
    #Main Artist
    artist_uri = track["track"]["artists"][0]["uri"]
    artist_info = sp.artist(artist_uri)
    
    #Name, popularity, genre
    artist_name = track["track"]["artists"][0]["name"]
    artist_pop = artist_info["popularity"]
    artist_genres = artist_info["genres"]
    
    #Album
    album = track["track"]["album"]["name"]
    
    #Popularity of the track
    track_pop = track["track"]["popularity"]
```

```python
# The caracteristics of the music that are available on each track URL
sp.audio_features(track_uri)[-1]
```

```python
len(sp.audio_features(track_uri))
```

```python
# Joining all the caracteristics of all the top 50 musics on a list (result)
list_range = list(range(0,50))
def join_musics(music):
  print(music)
  return sp.audio_features(track_uri)[music]

result = list(map(join_musics, list_range))
result
```

```python
# Making a Pandas DataFrame out of the list that we made on the step before (result) with each music in a row and the caracteristics/variables on the columns
spotipy_musics = pd.DataFrame(result)
spotipy_musics.head()
```

```python
# Changing again the duration column name
spotipy_musics.rename(columns={'duration_ms': 'duration'},
          inplace=True, errors='raise')
```

```python
# Passing the duration to seconds and rounding it
spotipy_musics['duration'] = spotipy_musics['duration'].apply(lambda time: round(time/1000))
spotipy_musics['duration'].head()
```

```python
 # Dropping unnecessary columns from the dataframe of the top 50 musics that we created
 spotipy_musics.drop(['type', 'mode', 'key'], axis = 1, inplace=True)
 spotipy_musics.columns
```

```python
 # Dropping unnecessary columns from the dataframe of the top 50 musics that we created
 spotipy_musics.drop(['uri','track_href','analysis_url'], axis = 1, inplace=True)
 spotipy_musics.columns
```

```python
 # Dropping unnecessary columns from the dataframe of the top 50 musics that we created
 spotipy_musics.drop(['tempo'], axis = 1, inplace=True)
 spotipy_musics.columns
```

```python
 # Trying to see some correlations between columns, if needed on the future
 spotipy_musics.corr()
```

```python
# Visualization of the correlations on a heatmap
plt.figure(figsize=(20,8))
sns.heatmap(spotipy_musics.corr(), annot=True, vmin=-1, vmax=1)
plt.title('Correlation on the Top 50 global musics')
plt.show()
```

```python
# Merging the spotify dataframe that we got from kaggle with the dataframe that we created from the top 50 musics playlist
spotify_bd = spotify[['track_id','artists','album_name','track_name','track_genre', 'popularity']]
final_df = spotify_bd.merge(spotipy_musics, left_on="track_id", right_on="id")
final_df.info()
```
